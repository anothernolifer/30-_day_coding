## Matplotlib example workflow

```python
# 1. Import matplotlib 

%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 3. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 4. Plot data
ax.plot(x,y)

# 5. Customize plot 
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 6. Save and show
fig.savefig("images/sample-plot.png")

```

# Plotting from pandas Dataframes

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].replace("[$,]", "", regex=True).astype(float)
car_sales
```

```python
car_sales["Sale date"] = pd.date_range("1/1/2025" , periods =len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Sale date" , y ="Total Sales");
```

```python
car_sales.plot(x="Sale date" , y ="Total Sales" , kind ="scatter");
```

```python
# * when plotting quickly , use pyplot method 
# * when plotting more advanced use matplotlib oo method
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# create histogram
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.plot.hist(figsize = (10,30),subplots = True);
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot method

over_50.plot(kind = 'scatter' ,
            x = 'age',
            y = 'chol',
            c = 'target');
```

```python
# oo method

fig , ax = plt.subplots(figsize =(10,6))

scatter = ax.scatter (x =over_50["age"],
                      y =over_50["chol"],
                      c = over_50["target"])

ax.set(title ="Heart disease and cholestrol levels",
       xlabel = "Age",
       ylabel = "Cholestrol");

ax.legend(*scatter.legend_elements(),title ="Target");

ax.axhline(over_50["chol"].mean(),
           linestyle = '--');
```

```python

fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 10))

scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')


ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol",
        xlim=[50, 80])  


ax0.axhline(y=over_50["chol"].mean(), 
            color="r", 
            linestyle="--", 
            label="Average");

ax0.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(y=over_50["chol"].mean(), 
            color="r", 
            linestyle="--");


scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')

ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate",
        ylim=[60, 220]) 


ax1.axhline(y=over_50["thalach"].mean(), 
            color="r", 
            linestyle="--");


ax1.legend(*scatter.legend_elements(), 
           title="Target")


fig.suptitle("Heart Disease Analysis", 
             fontsize=16, 
             fontweight="bold");
```
