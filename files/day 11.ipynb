# creating training data to form exponential growth shape

```python


torch.manual_seed(0)                      # fixes random state
observations = 2048
train_data = torch.zeros((observations , 2))  # creates tensor with 2048 rows and 2 columns
train_data[:,0]=50*torch.rand(observations)   # generates value of x between 0 and 50
train_data[:,1]=1.08**train_data[:,0]         # generates values of y =1.08^x 
```

## visualizing relation between x and y

```python
import matplotlib.pyplot as plt


fig=plt.figure(dpi=100,figsize=(8,6))
plt.plot(train_data[:,0],train_data[:,1],".",c="r")     # plot relation between x and y
plt.xlabel("values of x",fontsize=15)
plt.ylabel("values of $y=1.08^x$",fontsize=15)          # labels y axis
plt.title("An exponential growth shape",fontsize=20)    # creates title
plt.show()
```

```python
# ex 3.1

import torch

torch.manual_seed(0)    
observations = 2048
train_data = torch.zeros((observations, 2))    
train_data[:,0]=10*(torch.rand(observations)-0.5)  
train_data[:,1]=torch.sin(train_data[:,0]) 
```

```python
# ex 3.2

fig=plt.figure(dpi=100,figsize=(8,6))
plt.plot(train_data[:,0],train_data[:,1],".",c="r")
plt.xlabel("values of x",fontsize=15)
plt.ylabel("values of $y=sin(x)$",fontsize=15)
plt.title("A sine shape",fontsize=20)
plt.show()
```

## preparing the training dataset 

```python
from torch.utils.data import DataLoader

batch_size=128
train_loader=DataLoader(
    train_data,
    batch_size=batch_size,
    shuffle=True)
```
